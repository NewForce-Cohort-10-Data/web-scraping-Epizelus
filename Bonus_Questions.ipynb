1. Navigate to https://www.billboard.com/charts/hot-100/. Using BeautifulSoup, extract out the This Week, artist, song, Last Week, Peak Position, and Weeks on Chart values into a pandas DataFrame. Hint: The HTML for the number one ranked song is slightly different from that of the rest of the songs.

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as BS

url = "https://www.billboard.com/charts/hot-100/"
response = requests.get(url)

# Print the status code
if response.status_code == 200:
    print("Response Code: " + str(response.status_code) + ". Success!" )
else:
    print("Response Code: " + str(response.status_code) + ". Failure!" )
```

```python
soup = BS(response.text)
```

```python
soup.findAll('h3', class_="c-title a-no-trucate a-font-primary-bold-s u-letter-spacing-0021 lrv-u-font-size-18@tablet lrv-u-font-size-16 u-line-height-125 u-line-height-normal@mobile-max a-truncate-ellipsis u-max-width-330 u-max-width-230@tablet-only", id='title-of-a-story')
```

```python
songTitles = [title.text.strip() for title in soup.findAll('h3', class_="c-title a-no-trucate a-font-primary-bold-s u-letter-spacing-0021 lrv-u-font-size-18@tablet lrv-u-font-size-16 u-line-height-125 u-line-height-normal@mobile-max a-truncate-ellipsis u-max-width-330 u-max-width-230@tablet-only", id='title-of-a-story')]
songTitles[:10]
```
